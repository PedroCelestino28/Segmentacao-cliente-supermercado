**Nesta etapa**, será feito além do pipeline construído no último notebook o **PCA** para a redução de dimensionalidade. O PCA é utilizado geralmente para análise de dados exploratória porque revela a estrutura interna dos dados e explica a variação nos dados, esta ferramenta trabalha analisando dados que contêm muitas variáveis.

```python
import pandas as pd
import os

os.environ["OMP_NUM_THREADS"] = "1"

DADOS = "../dados/Mall_Customers_no_CustomerID.csv"

df = pd.read_csv(DADOS)
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

preprocessing = ColumnTransformer(
    [
        ("standard", StandardScaler(), ["Age", "Annual Income (k$)", "Spending Score (1-100)"]),
        ("onehot", OneHotEncoder(), ["Gender"])
    ]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

RANDOM_STATE = 42

pipeline = Pipeline(
    [
        ("preprocessing", preprocessing),
        ("pca", PCA(n_components=3, random_state=RANDOM_STATE)),
        ("clustering", KMeans(n_clusters=5, random_state=RANDOM_STATE, n_init=10))
    ]
)

pipeline.fit(df)
```

```python
import joblib

joblib.dump(pipeline, "../modelos/pipeline_preprocessing_pca_clustering.pkl")
```

**Esse projeto** projeto consiste na análise e identificação da melhor probabilidade de segmentação ou clusterização da base de dados de um supermercado.

Inicialmente no projeto, foi feito uma análise exploratória da base de dados de um supermercado, sendo verificado inicialmente através do *ydata-profiling* que gerou um report, onde a quantidade de clusters mais adequado para se trabalhar no projeto foi identificada. Após a análise do report, o arquivo foi armazenado em pasta identificada de modo a segmentar por pasta e arquivos as etapas as quais o projeto se dedicou a seguir.

Com as probabilidades de clusters definidos, foi construído um pipeline que teve como finalidade a implementação de um pre processamento dos dados. Sendo utilizado para testes de confirmação de números mais adequado de clusters o método do cotovelo e método da silhueta, sendo a opção definida de 4 clusters. Com isso, foi verificado através de gráficos de dispersão a distribuição dos dados e se os clusters ficariam distintos entre si. Além da distruição, foi verificado também os gráficos em boxplot para comparativo entre máximos, minimos e média de cada clusters selecionado dentre os parâmetros de idade, renda anual e score de gastos. Foi feita essa mesma comparação separando os gêneros para verificar se esse parâmetro gerava algum impacto significativo nas análises.

Por fim, foi inserido na parte final do projeto um PCA (principal component analysis) para sintetizar a maior parte das informações e manter mesmo assim a finalidade principal da análise garantindo a consistência da escolha do número de clusters e mostrando a sua distinção entre si.
